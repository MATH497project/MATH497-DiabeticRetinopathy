```python
path=os.getcwd()+'/data/'
```

### Exploartory data visuals

```python
person_profile_df=pd.read_pickle(path+'person_profile_df.pickle')
person_profile_df.head()
```

```python
person_profile_df.Age.plot.hist(bins=50)
plt.show()
```

```python
avg=person_profile_df.Enc_list_count.mean()
plt.hist([i for i in person_profile_df.Enc_list_count if i<avg])
plt.show()
```

```python
sizes=[int(len([i for i in person_profile_df.Enc_list_count if i<=avg])*100/len(person_profile_df.Enc_list_count)), 
       int(len([i for i in person_profile_df.Enc_list_count if i>avg])*100/len(person_profile_df.Enc_list_count))]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=['#encounter<=5', '#encounter>5'], autopct='%1.1f%%')
ax1.axis('equal') 
plt.show()
```
